```python
cf.index = cf.timestamp
```

```python
cf.drop(['timestamp'], axis =1, inplace = True)
```

```python
cf
```

```python
date_new = pd.date_range(cf.index[0],cf.index[-1], freq='D',normalize=True)
```

```python
date_new
```

```python
new_index = date_new.union(cf.index)
```

```python
new_cf = cf.reindex(new_index)
new_cf.head()
```

```python
new_cf = new_cf.interpolate(method='time')
new_cf.weekdays = new_cf.index.weekday_name
```

```python
new_cf = new_cf.resample('D').asfreq()
```

```python
new_cf.drop('contributor',axis =1, inplace = True)
```

```python
new_cf.head()
```

```python
new_cf.plot(style='-', figsize=(20,4))
```

```python
cf_day = new_cf.groupby('weekdays').mean()['coffees_made_today']
cf_day.plot(kind='bar')
```

```python
cf_made = new_cf.coffees.diff().shift(-1)
```

```python
cf_day['coffees_made_today']=cf_made
```
